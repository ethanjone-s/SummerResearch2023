## Plot Filament Spines

```python
# all_filament_spines.fits did not have the names correspoding to the right filaments,
# so I read in each .fits file

virgoIII=Table.read(cwd+'spines/filament_spine_VirgoIII.fits')

ursamajor=Table.read(cwd+'spines/filament_spine_Ursa_Major_Cloud.fits')

serpens=Table.read(cwd+'spines/filament_spine_Serpens.fits')

ngc5353=Table.read(cwd+'spines/filament_spine_NGC5353_4.fits')

leoIIB=Table.read(cwd+'spines/filament_spine_LeoII_B.fits')

leoIIA=Table.read(cwd+'spines/filament_spine_LeoII_A.fits')

leominor=Table.read(cwd+'spines/filament_spine_Leo_Minor.fits')

leominorB=Table.read(cwd+'spines/filament_spine_Leo_Minor_B.fits')

draco=Table.read(cwd+'spines/filament_spine_Draco.fits')

coma=Table.read(cwd+'spines/filament_spine_Coma_Berenices.fits')

canes=Table.read(cwd+'spines/filament_spine_Canes_Venatici.fits')

canesB=Table.read(cwd+'spines/filament_spine_Canes_B.fits')
```

```python
# created names array for each filament for legend later
names=['Virgo III','Ursa Major','Serpens','NGC5353','Leo IIB','Leo IIA',
      'Leo Minor','Leo Minor B','Draco','Coma Berenices','Canes Venatici','Canes B']

colors=['red','blue','yellow','green','purple','black',
        'brown','orange','pink','teal','gray']

# compiled list of RAs for each filament in one master array
RAs=[[virgoIII['ra']],[ursamajor['ra']],[serpens['ra']],[ngc5353['ra']],[leoIIB['ra']],[leoIIA['ra']],
     [leominor['ra']],[leominorB['ra']],[draco['ra']],[coma['ra']],[canes['ra']],[canesB['ra']]]

# same as RA; master array of compiled DECs
DECs=[[virgoIII['dec']],[ursamajor['dec']],[serpens['dec']],[ngc5353['dec']],[leoIIB['dec']],[leoIIA['dec']],
     [leominor['dec']],[leominorB['dec']],[draco['dec']],[coma['dec']],[canes['dec']],[canesB['dec']]]
```

```python
plt.figure()
for i in range(0,11):
    plt.scatter(RAs[i],DECs[i],c=colors[i],s=10,label=f'{names[i]}')
plt.xlabel('RA',fontsize=11); plt.ylabel('DEC',fontsize=11); plt.title('Plotting Locations of Filament Spines',fontsize=13)
plt.legend(fontsize=6)
```
